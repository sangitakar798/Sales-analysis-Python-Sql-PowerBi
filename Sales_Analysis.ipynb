```python
import pandas as pd
import os
```

```python
df = pd.read_csv("E:/SalesAnalysisExcelPythonSqlPowerBi/sales_raw_500.csv")
```

```python
df.head()
```

```python
#duplicate removes
df = df.drop_duplicates()
```

```python
# fix date column
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors = 'coerce')
```

```python
df = df.dropna(subset=['Order_Date'])
```

```python
df
```

```python
# sale - cost = profit
# from order_date (separate year and month)
```

```python
df['Profit'] = df['Sales'] - df['Cost']
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
df
```

```python
df.to_csv('E:/SalesAnalysisExcelPythonSqlPowerBi/sales_cleaned.csv',index=False)
```
